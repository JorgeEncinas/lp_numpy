# Exercise: Generate and Plot the XOR Dataset
What we'll do is if the value is bigger than 0 we count it as 1. Otherwise count it as 0.

Then apply the labeling with XOR logic.

```python
def XOR(sample):
  a, b = sample[0], sample[1]
  #print(a, b)
  if (a > 0 and b > 0):
    return 0
  elif (a > 0 or b > 0):
    return 1
  return 0
```

```python
x = np.random.randn(10000)
x2 = np.random.randn(10000)
X = np.vstack((x, x2))
```

```python
Y = np.apply_along_axis(XOR, axis=0, arr=X)
Y.shape
```

```python
plt.scatter(X[0, :], X[1, :], c=Y)
```

```python
# Since it's a normal distribution in the last one, you can see how values farther away are less likely.
# Let's try with a uniform distribution. It goes between one and zero, so we'll have to change it to go
# between -1 and 1
x = np.random.random(10000)
x2 = np.random.random(10000)
x -= 0.5
x *= 2
x2 -= 0.5
x2 *= 2
x.shape, x2.shape
```

```python
X = np.vstack((x, x2))
X
plt.scatter(X[0, :], X[1, :])
```

```python
# Good, now let's color this
Y = np.apply_along_axis(XOR, axis=0, arr=X)
Y.shape
```

```python
plt.scatter(X[0, :], X[1, :], c=Y)
```
